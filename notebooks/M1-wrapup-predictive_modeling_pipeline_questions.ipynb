```python
import pandas as pd
ames_housing = pd.read_csv("../datasets/house_prices.csv", na_values="?")
ames_housing = ames_housing.drop(columns="Id")

target_name = "SalePrice"
data, target = ames_housing.drop(columns=target_name), ames_housing[target_name]
target = (target > 200_000).astype(int)
```

Numerical features :

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate

numerical_features = [
  "LotFrontage", "LotArea", "MasVnrArea", "BsmtFinSF1", "BsmtFinSF2",
  "BsmtUnfSF", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "LowQualFinSF",
  "GrLivArea", "BedroomAbvGr", "KitchenAbvGr", "TotRmsAbvGrd", "Fireplaces",
  "GarageCars", "GarageArea", "WoodDeckSF", "OpenPorchSF", "EnclosedPorch",
  "3SsnPorch", "ScreenPorch", "PoolArea", "MiscVal",
]
data_numeric = data[numerical_features]
model = make_pipeline(StandardScaler(), 
                      SimpleImputer(strategy = "mean"), 
                      LogisticRegression())
cv_results1 = cross_validate(model, data_numeric, target, cv = 5)
scores = cv_results1['test_score']
print(scores.mean())
```

Add categorical features :

```python
categorical_features = list(set(data.columns) - set(numerical_features))
print(len(numerical_features))
print(len(categorical_features))
print(len(data.columns))
print(categorical_features)
```

On teste l'encodage OneHot sans handle unknown :

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Cat.
enc = OneHotEncoder()
enc.fit(data[categorical_features])
print(enc.categories_)
# print(enc.get_feature_names())
data_enc = enc.transform(data[categorical_features])
# print(data_enc)
```

Le souci `handle_unknown` n'apparaît pas lorsqu'on encode seulement, mais lorsqu'on entraîne également le modèle, car lors des cross validations, certaines catégories peuvent être dans le lot de test mais pas dans le lot d'entraînement

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Cat.
onehot_imputer_transformer = make_pipeline(SimpleImputer(strategy = "most_frequent"), OneHotEncoder(handle_unknown="ignore"))

# Num.
scaler_imputer_transformer = make_pipeline(StandardScaler(), SimpleImputer(strategy = "mean"))

# Preproc.
preprocessor = ColumnTransformer(transformers=[
    ("cat-preprocessor", onehot_imputer_transformer, categorical_features),
    ("num-preprocessor", scaler_imputer_transformer, numerical_features)
],
    remainder="passthrough")

# Model
model = make_pipeline(preprocessor, LogisticRegression(max_iter = 500))
cv_results2 = cross_validate(model, data, target, cv = 5)
print(cv_results2['test_score'].mean())
```

Does the model improve when adding categorical variables ?

```python
print((cv_results2['test_score'].mean() - cv_results1['test_score'].mean()) > (3 * cv_results2['test_score'].std()))
```
